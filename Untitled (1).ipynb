```python
import os
import faiss
import numpy as np
from pypdf import PdfReader
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
def load_documents(folder_path):

    docs = []

    for file in os.listdir(folder_path):

        if file.endswith(".pdf"):

            reader = PdfReader(os.path.join(folder_path, file))

            for page in reader.pages:
                text = page.extract_text()
                docs.append(text)

    return docs
```

```python
def build_index(folder_path):

    docs = load_documents(folder_path)

    embeddings = model.encode(docs)

    dimension = embeddings.shape[1]

    index = faiss.IndexFlatL2(dimension)

    index.add(np.array(embeddings))

    faiss.write_index(index, "restaurant.index")

    return docs
```

```python
def load_index():

    index = faiss.read_index("restaurant.index")

    docs = load_documents(folder_path)

    return index, docs
```

```python
def search(query, index, docs):

    query_vector = model.encode([query])

    distances, indices = index.search(np.array(query_vector), k=3)

    results = [docs[i] for i in indices[0]]

    return results
```

```python
folder_path = r"C:\Users\syedk\Downloads\restaurant_bot_documents"

if not os.path.exists("restaurant.index"):

    print("Creating restaurant index...")
    docs = build_index(folder_path)

else:

    print("Loading restaurant index...")

index, docs = load_index()

print("\nRestaurant AI Assistant Ready 🍽️")

while True:

    query = input("\nAsk about restaurant: ")

    results = search(query, index, docs)

    for r in results:
        print("\nAnswer:", r)
```
